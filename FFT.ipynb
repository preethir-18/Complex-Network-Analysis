```python
import numpy as np
from matplotlib import pyplot as plt

SAMPLE_RATE = 500  # Hertz For every second, 500 data points are calculated
DURATION = 1  # Seconds-Done for 1 second

def generate_sine_wave(freq, sample_rate, duration):
    x = np.linspace(0, duration, sample_rate * duration, endpoint=False)
    frequencies = x * freq 
    y = np.sin((2 * np.pi) * frequencies) #sin(2*pi*f)-angular frequency
    return x, y

x, y = generate_sine_wave(10, SAMPLE_RATE, DURATION)
u, v = generate_sine_wave(20, SAMPLE_RATE, DURATION)
z=y+v
plt.plot(x, y)
plt.xlabel('Time')
plt.ylabel('Voltage')
plt.title("10 Hz Signal")
plt.show()
plt.plot(u,v)
plt.xlabel('Time')
plt.ylabel('Voltage')
plt.title("20 Hz Signal")
plt.show()
plt.plot(z)
plt.xlabel('Time')
plt.ylabel('Voltage')
plt.title(" Sum of 10 Hz and 20 Hz Signal")
plt.show()

from scipy.fft import fft, fftfreq

N = SAMPLE_RATE * DURATION

yf = fft(z)
xf = fftfreq(N, 1 / SAMPLE_RATE)

plt.plot(xf, np.abs(yf))
plt.xlabel('Frequency')
plt.ylabel('Magnitude')
plt.title("FFT on Combined Signal")
plt.show()
```
